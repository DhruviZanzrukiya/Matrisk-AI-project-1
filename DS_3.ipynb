Bridge Age Distribution


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['age_years'], bins=25)
plt.title('Distribution of Bridge Age')
plt.xlabel('Age (Years)')
plt.ylabel('Number of Bridges')
plt.show()
```

Corrosion Environment Comparison


```python

import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(
    x='corrosion_environment',
    y='corrosion_rate_mm_yr',
    data=df
)
plt.title('Corrosion Rate by Environment')
plt.show()
```

Age vs Condition Rating


```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='age_years',
    y='condition_rating',
    data=df
)
plt.title('Age vs Condition Rating')
plt.show()
```

Material Performance Comparison


```python
plt.figure(figsize=(12,6))
sns.barplot(
    x=df.groupby('material')['condition_rating'].mean().index,
    y=df.groupby('material')['condition_rating'].mean().values
)

plt.xticks(rotation=60)
plt.title('Average Condition Rating by Material')
plt.show()
```

Structural Deficiency Analysis


```python
plt.figure(figsize=(6,5))
sns.boxplot(
    x='structurally_deficient',
    y='corrosion_rate_mm_yr',
    data=df
)

plt.title('Corrosion Rate vs Structural Deficiency')
plt.show()
```

Thickness Loss Distribution


```python
df['loss_mm'] = (
    df['original_thickness_mm']
    - df['remaining_thickness_mm']
)

plt.figure(figsize=(8,5))
plt.hist(df['loss_mm'], bins=30)
plt.title('Thickness Loss Distribution')
plt.xlabel('Thickness Loss (mm)')
plt.show()
```

Remaining Life Analysis


```python
df['remaining_life'] = (
    df['design_life_years']
    - df['age_years']
)

plt.figure(figsize=(8,5))
plt.hist(df['remaining_life'], bins=30)
plt.title('Remaining Life Distribution')
plt.xlabel('Years Remaining')
plt.show()
```

Financial Exposure Analysis


```python

plt.figure(figsize=(8,5))
plt.hist(df['replacement_cost_M'], bins=30)
plt.title('Replacement Cost Distribution')
plt.xlabel('Replacement Cost (Million $)')
plt.show()
```

Loan Exposure by Bridge Type


```python
loan_data = df.groupby(
    'bridge_type'
)['loan_outstanding_M'].sum()

loan_data.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Total Loan Outstanding by Bridge Type')
plt.ylabel('Loan (Million $)')
plt.show()
```

Traffic Exposure Analysis


```python
traffic = df.groupby(
    'bridge_type'
)['adt'].mean()

traffic.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Average Daily Traffic by Bridge Type')
plt.ylabel('ADT')
plt.show()
```

Correlation Heatmap 

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.select_dtypes(include=['int64','float64']).corr(),
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Matrix')
plt.show()
```

MatRisk AI Risk Score 

```python
df['risk_score'] = (
    df['corrosion_rate_mm_yr']*100
    + df['structurally_deficient']*20
    + (10-df['condition_rating'])*5
)

plt.figure(figsize=(8,5))
plt.hist(df['risk_score'], bins=30)
plt.title('Bridge Risk Score Distribution')
plt.xlabel('Risk Score')
plt.show()
```
